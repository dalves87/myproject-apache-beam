Carregar o Dataset
Primeiro, carregamos o dataset em um DataFrame usando a biblioteca Pandas.

```python
import pandas as pd

# Carregar o dataset
dataset_df = pd.read_csv('integrated_output-00000-of-00001.csv', delimiter=';')
```

Análise Exploratória dos Dados
1. Resumo Estatístico:
Vamos começar com um resumo estatístico das notas.

```python
# Resumo estatístico das notas
print(dataset_df[['Nota_Matematica', 'Nota_Portugues', 'Nota_Ciencias']].describe())
```

2. Distribuição das Notas por Matéria:
Criaremos histogramas para visualizar a distribuição das notas em Matemática, Português e Ciências.

```python
import matplotlib.pyplot as plt

# Histograma das notas de Matemática
plt.figure(figsize=(10, 6))
plt.hist(dataset_df['Nota_Matematica'], bins=10, color='blue', alpha=0.7, rwidth=0.85)
plt.title('Distribuição das Notas de Matemática')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()

# Histograma das notas de Português
plt.figure(figsize=(10, 6))
plt.hist(dataset_df['Nota_Portugues'], bins=10, color='green', alpha=0.7, rwidth=0.85)
plt.title('Distribuição das Notas de Português')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()

# Histograma das notas de Ciências
plt.figure(figsize=(10, 6))
plt.hist(dataset_df['Nota_Ciencias'], bins=10, color='red', alpha=0.7, rwidth=0.85)
plt.title('Distribuição das Notas de Ciências')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()
```

3. Correlação entre Notas:
Vamos verificar a correlação entre as notas das diferentes disciplinas para identificar possíveis padrões.

```python
# Matriz de correlação
correlacao = dataset_df[['Nota_Matematica', 'Nota_Portugues', 'Nota_Ciencias']].corr()

print(correlacao)

# Mapa de calor para visualizar a correlação
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlação entre as Notas')
plt.show()
```

4. Análise das Preferências Alimentares:
Vamos analisar a distribuição das preferências alimentares entre os alunos.

```python
# Contagem das preferências
preferencias_contagem = dataset_df[['Gosta_Frutas', 'Gosta_Vegetais', 'Gosta_Doces']].apply(pd.Series.value_counts)

# Gráfico de barras para visualizar as preferências
preferencias_contagem.plot(kind='bar', figsize=(10, 6))
plt.title('Distribuição das Preferências Alimentares')
plt.ylabel('Número de Alunos')
plt.xlabel('Preferências')
plt.show()
```

Análises Conjuntas
Podemos agora realizar algumas análises para ver se há alguma relação entre o desempenho acadêmico e as preferências alimentares.

1. Relação entre Nota e Preferência por Frutas:

```python
# Gráfico de dispersão: relação entre nota de Matemática e gostar de frutas
sns.boxplot(x='Gosta_Frutas', y='Nota_Matematica', data=dataset_df)
plt.title('Nota de Matemática vs. Preferência por Frutas')
plt.show()

# Gráfico de dispersão: relação entre nota de Português e gostar de frutas
sns.boxplot(x='Gosta_Frutas', y='Nota_Portugues', data=dataset_df)
plt.title('Nota de Português vs. Preferência por Frutas')
plt.show()
```

2. Relação entre Nota e Preferência por Vegetais:

```python
# Gráfico de dispersão: relação entre nota de Ciências e gostar de vegetais
sns.boxplot(x='Gosta_Vegetais', y='Nota_Ciencias', data=dataset_df)
plt.title('Nota de Ciências vs. Preferência por Vegetais')
plt.show()
```
